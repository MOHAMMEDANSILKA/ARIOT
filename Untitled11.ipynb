```python
import numpy as np
import math
import matplotlib.pyplot as plot

# FSK modulation
amplitude = 1
frequencies = [1200, 4000]
X = []
Y1 = []
Y2 = []
signal = [1, 0, 1, 1, 0, 1, 0, 1, 0, 0]
interval = 1000
phase = 0

for s in signal:
    frequency = frequencies[s]
    for t in range(interval):
        phase += frequency / 1000000 * math.pi * 2
        Y2.append(s)
        Y1.append(math.sin(phase) * amplitude)

X = range(len(signal) * interval)
ax, fig = plot.subplots(figsize=(16, 2))
plot.plot(X, Y1)
plot.plot(X, Y2)
plot.title('Digital Signal')
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.grid(True, which='both')
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# QAM modulation
amplitude = 1
frequencies = [1200, 4000]
symbols = [0, 1]
X = []
Y1 = []
Y2 = []
signal = [symbols[i] for i in [1, 0, 1, 0, 0, 1, 0, 1, 0, 0]]  # Adjusted the signal to use 0 and 1
interval = 1000
phase = 0

for s in signal:
    frequency = frequencies[s]
    for t in range(interval):
        phase += frequency / 1000000 * math.pi * 2
        Y2.append(s)
        Y1.append(math.sin(phase) * amplitude)

X = range(len(signal) * interval)
fig, ax = plt.subplots(figsize=(16, 2))
plt.plot(X, Y1)
plt.plot(X, Y2)
plt.title('Quadrature Amplitude Modulation (QAM)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# QPSK modulation
amplitude = 1
frequencies = [0, np.pi / 2, np.pi, -np.pi / 2]  # Four phase shifts (0, 90, 180, -90 degrees)
symbols = [0b00, 0b01, 0b10, 0b11]  # Two bits per symbol
X = []
Y1 = []
Y2 = []
signal = [symbols[i] for i in [0, 2, 1, 3, 2, 0, 1, 3]]  # Example signal data
interval = 1000
phase = 0

for s in signal:
    phase_shift = frequencies[s]
    for t in range(interval):
        phase += phase_shift
        Y2.append(s)
        Y1.append(math.cos(phase) * amplitude)  # In-phase component

X = range(len(signal) * interval)
fig, ax = plt.subplots(figsize=(16, 2))
plt.plot(X, Y1)
plt.plot(X, Y2)
plt.title('Quadrature Phase Shift Keying (QPSK),ansil')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.show()
```
